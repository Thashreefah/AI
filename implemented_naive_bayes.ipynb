```python
#Naive baiyes implementation code
import numpy as nm  
import matplotlib.pyplot as plt 
import pandas as pd  
import numpy as np

```

```python
# Importing the dataset  
dataset = pd.read_csv('User_data.csv')  
x = dataset.iloc[:, [2, 3]].values  
y = dataset.iloc[:, 4].values  
x
```

```python
print(dataset)
```

```python
print(y)
```

```python
# Splitting the dataset into the Training set and Test set  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)  
  
```

```python
print(x_train)
```

```python
# Feature Scaling 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()  
x_train1 = sc.fit_transform(x_train)  
x_test1 = sc.transform(x_test)  
```

```python
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
x_train2 = sc.fit_transform(x_train)  
x_test2 = sc.transform(x_test)  
```

```python
print(x_train1)
```

```python
print(x_train2)
```

```python
#convert array values to datafram
df = pd.DataFrame(x_train1)               
```

```python
df
```

```python
df1 = pd.DataFrame(x_train2)
```

```python
df1
```

```python
#Concatenate the two dataframes
df2=pd.concat([df,df1],axis=1)
df2
```

```python
#Convert arrays into dataframes and concatenate the dataframes
dff=pd.concat([pd.DataFrame(x_test1),pd.DataFrame(x_test2)],axis=1)
dff
```

```python
# Fitting Naive Bayes to the Training set  
from sklearn.naive_bayes import GaussianNB  
classifier = GaussianNB()  
classifier.fit(df2,y_train)  
```

```python
# Predicting the Test set results  
y_pred = classifier.predict(dff)  
print(y_pred)
```

```python
# Making the Confusion Matrix  
from sklearn.metrics import confusion_matrix  
cm = confusion_matrix(y_test, y_pred) 
print('Confusion Matrix : \n',cm)
```

```python
from sklearn.metrics import accuracy_score
print ('Accuracy : ', accuracy_score(y_test, y_pred))
```

```python
from sklearn import metrics
import numpy as np
y_pred=np.around(classifier.predict(dff))
print(metrics.classification_report(y_test,y_pred))
```
